```python
import os
import dotenv
import argparse

import numpy as np
import pandas as pd

from maldi_learn.driams import load_driams_dataset

dotenv.load_dotenv()
DRIAMS_ROOT = os.getenv('DRIAMS_ROOT')
```

```python
#strat_ids = ['PATIENTENNUMMER_id', 'FALLNUMMER_id', 'AUFTRAGSNUMMER_id']
col_strat_id = 'FALLNUMMER_id'
col_strat_count = col_strat_id + '_n_duplicates'
```

```python
# species = 'Escherichia coli'
species = 'Staphylococcus aureus'
# antibiotic = 'Ceftriaxone'
antibiotic = 'Oxacillin'

filename = './match_strat_to_code.csv'
```

```python
# extract unique stratification ids
df = pd.read_csv(filename, low_memory='False')
df_duplicates = df.loc[
                df[col_strat_count] > 1
                       ]
duplicated_strat_ids = df_duplicates[
                                col_strat_id
                                     ].unique()
```

```python
# create dict to match strat_id with Bruker codes
duplicate_dict = {}
for strat_id in duplicated_strat_ids:
    df_ = df.loc[df[col_strat_id] == strat_id]
    duplicate_dict[strat_id] = list(df_['code'].values)
```

```python
# load DRIAMS
driams_dataset = load_driams_dataset(
            DRIAMS_ROOT,
            'DRIAMS-A',
            '*',
            species,
            antibiotic,
            spectra_type='binned_6000',
            #nrows=1000, #FIXME
        )
codes_in_X = driams_dataset.y.code.values
```

```python
for strat_id in duplicated_strat_ids:
    year = strat_id.split('_')[1]
    
    for code in duplicate_dict[strat_id]:
        if sum(codes_in_X == code) != 0:
            strat_id_sane = strat_id.replace(' ', '_')
            original_file = f'/links/groups/borgwardt/Data/DRIAMS/DRIAMS-A/binned_6000/{year}/{code}.txt'
            cp_path = f'/links/groups/borgwardt/Data/DRIAMS_strat/DRIAMS-A/binned_6000/{year}/{strat_id_sane}/'
            cp_file = f'/links/groups/borgwardt/Data/DRIAMS_strat/DRIAMS-A/binned_6000/{year}/{strat_id_sane}/{code}.txt'
            print(original_file, cp_path)
            !echo $cp_path
            !mkdir $cp_path
            !cp $original_file $cp_file
```
