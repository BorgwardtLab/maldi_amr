## Define functions

```python
def count_elements(array):
    counts = np.unique(array, return_counts=True)
    zipped = zip(counts[0],counts[1])
    return sorted(zipped, key=operator.itemgetter(1), reverse=True)

def intersect_header_all_sheets(book, nsheets):
    for i in range(nsheets):
        sheet = book.sheet_by_index(i)
        ncols = sheet.ncols
        nrows = sheet.nrows

        header = [sheet.cell_value(0,c).encode('utf-8') for c in range(ncols)]
        if i==0:
            header_intersect_sheets = set(header)
        else:
            header_intersect_sheets = header_intersect_sheets.intersection(set(header))
    return header_intersect_sheets
    
def count_values_per_key(dictionary):
    list_len = []
    for key in dictionary.keys():
        list_len.append(len(dictionary[key]))
    return np.unique(list_len, return_counts=True)

def del_keys_more_values_than(dictionary, max_num_values, verbose=0):
    for key in dictionary.keys():
        length = len(dictionary[key])
        if length > max_num_values:
            if verbose > 0:
                print('{}, {}'.format(key, dictionary[key]))
            del dictionary[key]
```

```python
def get_dict_from_Viollier_IDRES(book,nsheets,verbose=0):
    # read in LabID and Isolate number for all sheets
    # make dictionary with all entries per sampleID
    header_union = intersect_header_all_sheets(book, nsheets)
    print 'intersected header: {}'.format(header_union)
    d_IDs = ddict(list)

    for i in range(nsheets):
        sheet = book.sheet_by_index(i)
        ncols = sheet.ncols
        nrows = sheet.nrows

        header = [sheet.cell_value(0,c).encode('utf-8') for c in range(ncols)]
        idx_LabID = header.index('Lab ID')
        idx_isolateNr = header.index('Isolate Number')
        idx_species = header.index('Organism Name')
        idx_antibiotics_start = header.index('AM-Ampicillin')
        list_antibiotics_all = header[idx_antibiotics_start:]
        list_antibiotics = [ab for ab in list_antibiotics_all if ab in header_union]

        for row_number in range(1,nrows):
            LabID = sheet.cell_value(row_number,idx_LabID)
            # valid LabIDs are only made up of digits
            if not isinstance(LabID, float):
                continue
            
            sampleID = str(int(LabID))
            if len(sampleID)!=10:
                if verbose==1: print('sampleID {} does not have 10 digits. continue.'.format(sampleID))
                continue

            species = str(sheet.cell_value(row_number,idx_species))
            antibiotics_all = [str(sheet.cell_value(row_number,i)) for i in range(idx_antibiotics_start,ncols)]
            antibiotics = [ab for i, ab in enumerate(antibiotics_all) if list_antibiotics_all[i] in header_union]
            d_IDs[sampleID].append({'species':species,'list_antibiotics':list_antibiotics,'MIC': antibiotics})
            assert len(antibiotics_all)==len(list_antibiotics_all)
            assert len(antibiotics)==len(list_antibiotics), 'List antibiotics and resistance phenotypes not the same length. {} {} {} {}'.format(len(antibiotics), len(list_antibiotics), antibiotics, list_antibiotics)
    
    if verbose==1: print('{} keys before del_keys_more_values_than.'.format(len(d_IDs.keys())))
    del_keys_more_values_than(d_IDs,max_num_values=1)
    if verbose==1: print('{} keys after del_keys_more_values_than.'.format(len(d_IDs.keys())))
    return d_IDs


def get_dict_TGNR_Bruker_Viollier(Bruker_LabID_match_fname, verbose=0):
    with open(Bruker_LabID_match_fname,'r') as f:
        ff = csv.reader(f, delimiter=';', dialect=csv.excel)

        d_BrukerIDs = ddict(list)
        non_10_counter = 0 
        
        for j, row in enumerate(ff):
            if j==0:
                idx_Bruker = 0
                idx_LabID = 1
                continue
            else:
                last_entry = row[idx_LabID].split('-')[-1]
                machine_number = row[idx_Bruker].split('_')[-1]
                BrukerID = row[idx_Bruker].split('_')[0]

                # define what part of the ID is used
                final_ID = last_entry[:10]
                
                if len(final_ID)!= 10:
                    non_10_counter+=1
#                     if verbose==1: print('final_ID {} does not have 10 digits. continue.'.format(final_ID))
                    continue
                d_BrukerIDs[final_ID].append(BrukerID +'_'+ machine_number)

    # only unique entries in d_BrukerIDs
    if verbose==1: print('{} keys before del_keys_more_values_than.'.format(len(d_BrukerIDs.keys())))
    del_keys_more_values_than(d_BrukerIDs,max_num_values=1)
    if verbose==1: print('{} keys after del_keys_more_values_than.'.format(len(d_BrukerIDs.keys())))
    if verbose==1: print('{} keys were deleted due to other lengths than 10.'.format(non_10_counter))
        
    print('{} keys extracted from Bruker file.'.format(len(d_BrukerIDs.keys())))
    return d_BrukerIDs


def count_matches_IDRES_Bruker(d_IDs, d_BrukerIDs):
    matched_IDs = []
    for LabID_key in d_IDs.keys():
        n_containing_substring = sum([LabID_key in spectraID_key for spectraID_key in d_BrukerIDs.keys()])
        print('\nThe LabID {} was found {} times among the spectraIDs'.format(LabID_key, n_containing_substring))

        if n_containing_substring == 1:
            matched_IDs.append(LabID_key)
    print '{} of LabIDs in the Viollier-dicts_code_TGNR.csv file matched with the Viollier_IDRES_clean.xlsx LabIDs'.format(len(matched_IDs))
    
    
def combine_IDRES_Bruker(d_IDs, d_BrukerIDs):
    d_combined = ddict(dict)

    for LabID_key in d_IDs.keys():
        n_containing_substring = sum([LabID_key in spectraID_key for spectraID_key in d_BrukerIDs.keys()])

        if n_containing_substring == 1:
            IDRES_dict = copy.deepcopy(d_IDs[LabID_key][0])
            BrukerID = d_BrukerIDs[LabID_key][0]
            assert id(IDRES_dict)!=id(d_IDs[LabID_key][0])
            IDRES_dict.update({'BrukerID':BrukerID})
            d_combined[LabID_key].update(IDRES_dict)
    return d_combined
```

```python
def get_MICmatrix_from_EUCAST(book, sheet_names):
    # create matrix with MIC for each species-antibiotic combination
    # rows = antibiotics
    # columns = species

    delete_strings = [' ',' ', '', 'Cephalosporins', 'Monobactams', 'Oxazolidinones','Fluoroquinolones','Glycopeptides','Penicillins',
                      'Macrolides','Aminoglycosides','Carbapenems', 'Glycopeptides and lipoglycopeptides', 'Macrolides, lincosamides and streptogramins',
                     'Tetracyclines', 'Oxazolidinones', 'Miscellaneous agents']
    ab_set = set()

    for sheet_name in sheet_names:
        sheet = book.sheet_by_name(sheet_name)
        ncols = sheet.ncols
        nrows = sheet.nrows

        content_ab = [sheet.cell_value(r,0).encode('utf-8').replace('\n',' ') for r in range(1,nrows)]
        content_ab_clean = [s for s in content_ab if s not in delete_strings]

        [ab_set.add(elem) for elem in content_ab_clean]

    ab_list = list(ab_set)
    ab_list.sort()

    zero_matrix = np.zeros((len(ab_list),len(sheet_names),2))
    ab_VS_species = np.full_like(zero_matrix, np.nan)
    
    # for sheet_name in species_sheet_names:
    for sheet_name in sheet_names:

        sheet = book.sheet_by_name(sheet_name)
        ncols = sheet.ncols
        nrows = sheet.nrows

        content_ab = [sheet.cell_value(r,0).encode('utf-8').replace('\n',' ') for r in range(nrows)]
    #     print content_ab

        for ab in ab_list:
            if ab in content_ab:
                idx = content_ab.index(ab)
                MIClow = sheet.cell_value(idx,1)
                MICupp = sheet.cell_value(idx,2)

                if type(MIClow)=='unicode':
                    MIClow.encode('utf-8')
                if type(MICupp)=='unicode':
                    MICupp.encode('utf-8')

                if type(MIClow)==unicode and MIClow.replace('.','',1).isdigit():
                    MIClow = float(MIClow)
                elif type(MIClow)==unicode and not MIClow.replace('.','',1).isdigit():
                    continue

                if type(MICupp)==unicode and MICupp.replace('.','',1).isdigit():
                    MICupp = float(MICupp)
                elif type(MICupp)==unicode and not MICupp.replace('.','',1).isdigit():
                    continue

                # add values to table
                ab_VS_species[ab_list.index(ab),sheet_names.index(sheet_name),0] = MIClow
                ab_VS_species[ab_list.index(ab),sheet_names.index(sheet_name),1] = MICupp
            else: 
                continue
    return ab_VS_species, ab_list, sheet_names
```

```python
def MIC_to_RSI(MIC, MIClow, MICupp):
    if MIC == '':
        return 'NaN'
    

    if np.isnan(MIClow) or np.isnan(MIClow):
        return 'NaN'
    assert isinstance(MIClow, float), 'MIClow not a float.'
    assert isinstance(MICupp, float), 'MICupp not a float.'
    
    # if MIC is just a float, w/o '<=>' signs
    if MIC.replace('.','',1).isdigit():
        mic = float(MIC)
        if mic <= MIClow:
            return 'S'
        elif mic >= MICupp:
            return 'R'
        elif mic <= MICupp and mic > MIClow:
            return 'I'
    
    # if MIC is 'less than' value
    if '<' in MIC:
        mic = MIC.replace('<','').replace('=','')  
        mic = float(mic.replace(',','.'))
        if mic <= MIClow:
            return 'S'
        else:
            return 'NaN'
        
    # if MIC is 'greater than' value
    if '>' in MIC:
        mic = MIC.replace('>','').replace('=','')  
        mic = float(mic.replace(',','.'))
        if mic > MICupp:
            return 'R'
        else:
            return 'NaN'
    return 'NaN'
```

```python
# Read in csv file
def convert(d_data):
    completeDataset_dict = ddict(list)

    keys = d_data.keys()
    for i, key in enumerate(keys):

        # initialize all antibiotics
        completeDataset_dict['sampleID'].append(d_data[key]['BrukerID'])
        completeDataset_dict['species'].append(d_data[key]['species'])
        completeDataset_dict['genus'].append(d_data[key]['species'].split(' ')[0])
        for j, ab in enumerate(d_data[key]['list_antibiotics']):
            completeDataset_dict[ab].append(d_data[key]['MIC'][j])
        
        x = [len(value) for _, value in completeDataset_dict.items()]
        assert len(np.unique(x)) == 1, '{} {}'.format(key,[(k, value) for k, value in completeDataset_dict.items()])

    # each value (e.g. each column in our dataframe) has the same length
    x = [len(value) for _, value in completeDataset_dict.items()]
    assert len(np.unique(x)) == 1

    Viollier_Dataset = pd.DataFrame.from_dict(completeDataset_dict)
    return Viollier_Dataset
```
